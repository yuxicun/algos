
### **相关题目**

[594. 字符串查找 II](https://www.lintcode.com/problem/strstr-ii)








### **解题代码**

````python
#由Amark 提供
def str_find(source_str, target_str):
    """
    查找源字符串中目标字符串出现的位置，未出现返回-1
    """
    i, j = 0, 0
    while i < len(source_str) - len(target_str):
        while j < len(target_str):
            if target_str[j] != source_str[j + i]:
                i += 1
                j = 0
            else:
                j += 1
        return i
    return -1

        
#由Tango提供
class Solution:
    """
    @param source: 
    @param target: 
    @return: return the index
    """
    def strStr(self, source, target):
        return (source.find(target))
````

### **解题代码**

````python
# Amark 提供
def is_minor_str(target_str):
    """
    给定⼀一个字符串串，判断其是否为⼀一个回⽂文串串。只考虑字⺟母和数字，忽略略⼤大⼩小写
    """
    # 空字符串直接返回False
    if not target_str:
        return False

    # 过滤字符串，只剩下字母与数字
    new_target_str = ''
    target_str = target_str.lower()
    for e in target_str:
        if e in string.ascii_lowercase or e in string.digits:
            new_target_str += e
    
    # 一个字母返回False
    target_str_len = len(new_target_str)
    if target_str_len == 1:
        return False
    
    # 开始遍历
    for i in range(target_str_len):
        # 如果不满足回文字符串条件直接返回False
        if new_target_str[i] != new_target_str[-i - 1]:
            return False
        # 直到满足下面条件的时候且上面没有拦截那必然是回文字符串
        if (i + 1) * 2 == target_str_len or (i + 1) * 2 + 1 == target_str_len:
            return True

        
# 由Tango 提供        
class Solution:
    """
    @param s: A string
    @return: Whether the string is a valid palindrome
    """
    def isPalindrome(self, s):
        # write your code here
        s_list = [i.lower() for i in s if i.isalnum()]
        if s_list==s_list[::-1]:
            return True
        else:
            return False
````


### **挑战**

时间复杂度O(*logn*)

### **解题代码**

````python
# 由Amark提供 方法一
def my_pow(num, n):
    """
    使用递归实现n次幂
    """
    if n < 0:
        num = 1 / num
        n = abs(n)
    if n == 0:
        return 1
    if n == 1:
        return num
    return my_pow(num, n - 1) * num
my_pow(2,-2)
​
# 由Amark提供方法二
def myPow(self, num, n):
        k = abs(n)
        new_num = 1
    
        if k == 0:
            return 1
    
        while k > 0:
            num_2 = num
            if n < 0:
                num_2 = 1 / num
            new_num = new_num * num_2
            k -= 1
        return new_num
my_pow(2,-2)


# 由Tango提供
def my_pow(number, n):
    judge = True
    if n < 0:
        n = -n
        judge = False
    if n == 0:
        return 1
    result = 1 
    count = 1
    temp = number
    while n > 0:
        if n >= count:
            result *= temp
            temp = temp * number
            n -= count
            count += 1
        else:
            temp /= number
            count -= 1
    return result if judge else 1/judge
````


## 9.快速幂「140」
https://www.lintcode.com/problem/fast-power/description

### **描述**


$$
计算a^n\%b，其中a,b 和n 都是32位的非负整数
$$


### **样例**

$$
例如 2^{31!}\%3=2
$$

$$
例如 100^{1000} \% 1000 = 0
$$



### **挑战**

O(logn)

### **相关题目**

[949. 斐波那契数列 II](https://www.lintcode.com/problem/fibonacci-ii)[428. x的n次幂](https://www.lintcode.com/problem/powx-n)




1.减小a的方法：

$$
积的取余等于每个数取余的乘积再取余
意思就是说，f(a, b, c) = a^{b!} MODc= (a MODc)^{b!}MODc
$$

$$
举个例子7^{8}\%2 = (7\%2)^8\%2 = 1^8\%2 = 1\%2 = 1
这个数学定理有效的帮助我们减小了a 的值。
$$


2.减小n的方法：
通过指数运算法则拆分n：

我们都知道指数的运算法则，!
$$
a^{x+y} = a^{x}*a^{y}
$$
 那么公式中的n！也可以通过拆分达到减小n的效果。



当b为偶数时：
$$
a^n=a^{n/2}*a^{n/2}=a^{n/2*2}=(a^2)^{n/2}
$$
当b为奇数时：
$$
a^n=a*a^{n-1}= a*a^{(n-1)/2}*a^{(n-1)/2}=a*a^{(n-1)/2*2}=a*(a^2)^{(n-1)/2}
$$




````python
# 计算输入的阶乘
def count_factorial(y):
    result = 1
    for each in range(1, y + 1):
        result *= each
    return result


def function(a, b, c):
    b = count_factorial(b)
    result = 1
    # 减小a
    a = a % c

    while b > 0:
        # 当b为奇数时
        if b % 2 == 1:
            result = (result * a) % c
        # b更新为b/2
        b = b >> 1
        # a更新为a^2 并 取余减小a的值
        a = (a ** 2) % c
    return result


print(function(2, 1, 2), 0)
print(function(3, 2, 2), 1)
print(function(1, 100000, 1), 0)
print(function(1, 100000, 2), 1)
print(function(99036, 92879, 77028), 0)
print(function(14916, 63624, 37968), 32544)
print(function(48778, 6070, 89146), 57188)
````

````python
#Amark 提供

quick_sort([3,2,1,4,5])
def quick_sort(nums):
    if len(nums) <= 1:
        return nums
    return quick_sort([l for l in nums[1:] if l < nums[0]]) + [nums[0]] + quick_sort([l for l in nums[1:] if l > nums[0]])
​
        
quick_sort([3,2,1,4,5])
        

# 快速排序，递归将大于key值和小于key值的分为两部分，直到每个子数组长度为1，不可再分。

class Solution:
    # @param {int[]} A an integer array
    # @return nothing
    def sortIntegers2(self, A):
        # Write your code here
        if A is None: return A
        n = len(A)
        if n == 1 or n == 0: return A
        return self.quickSort(A,0,n-1)
         
    def quickSort(self,List,left,right):
        if left >= right:
            return List
        key = List[left]
        low = left
        high = right
        while left < right:
            while left < right and List[right] >= key:
                right -= 1
            List[left],List[right] = List[right],List[left]
            while left < right and List[left] <= key:
                left += 1
            List[right],List[left] = List[left],List[right]
        self.quickSort(List,low,left-1)
        self.quickSort(List,left+1,high)
        return List
````

````python
# 由Amark提供
def two_sum(nums, target):
    """
    给定一个已经 按升序排列 的数组，找到两个数使他们加起来的和等于特定数
    """
    from collections import defaultdict
    
    nums_dict = defaultdict(dict)
    for i, e in enumerate(nums):
        if nums_dict[e]:
            return [nums_dict[e]['soulmate'], i]
        
        find_num = target - e
        nums_dict[find_num] = {"soulmate": i}
        



class Solution:
    """
    @param nums: an array of Integer
    @param target: target = nums[index1] + nums[index2]
    @return: [index1 + 1, index2 + 1] (index1 < index2)
    """
    def twoSum(self, nums, target):
        # write your code here
        #双指针写法
        length = len(nums)
        left, right = 0, length - 1
        
        while left < right:
            sum = nums[left] + nums[right] 
            if sum == target:
                return [left + 1, right + 1]
            
            else:
                if sum > target:
                    right -= 1 
                else:
                    left += 1 
            
````

Given a linked list, return the node where the cycle begins. If there is no cycle, return `null`.

To represent a cycle in the given linked list, we use an integer `pos` which represents the position (0-indexed) in the linked list where tail connects to. If `pos` is `-1`, then there is no cycle in the linked list.

**Note:** Do not modify the linked list.

 

**Example 1:**

```
Input: head = [3,2,0,-4], pos = 1
Output: tail connects to node index 1
Explanation: There is a cycle in the linked list, where tail connects to the second node.
```

**Example 2:**

```
Input: head = [1,2], pos = 0
Output: tail connects to node index 0
Explanation: There is a cycle in the linked list, where tail connects to the first node.
```

![img](https://assets.leetcode.com/uploads/2018/12/07/circularlinkedlist_test2.png)

**Example 3:**

```
Input: head = [1], pos = -1
Output: no cycle
Explanation: There is no cycle in the linked list.
```

![img](https://assets.leetcode.com/uploads/2018/12/07/circularlinkedlist_test3.png)

````python
# 由Tango提供

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self,head):
        if head !=None:
            d = {}
            while head and head.next:
                if head in d:
                    return head
                else:
                    d[head] = head
                head = head.next
        else:
            return None
            
        
````
